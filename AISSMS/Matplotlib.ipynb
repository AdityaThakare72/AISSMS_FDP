```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.xlabel('First Value')
plt.ylabel('Second Value')
# markers: >, <, ^, o
plt.grid()
plt.plot(x, y, marker = 'o', mfc = 'r', ls = '--');
```
